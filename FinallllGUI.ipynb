```python
from transformers import MarianTokenizer, MarianMTModel
import torch
import tkinter as tk


saved_model_directory = r"E:\course\lvl4\fine-tuned-en-ar-20250501T123312Z-001\fine-tuned-en-ar"


tokenizer = MarianTokenizer.from_pretrained(saved_model_directory)
model = MarianMTModel.from_pretrained(saved_model_directory)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
model.eval()


def translate():
    sentence = input_text.get("1.0", tk.END).strip()
    if not sentence:
        output_text.delete("1.0", tk.END)
        output_text.insert(tk.END, "Please enter a sentence.")
        return

    # Tokenize and translate
    inputs = tokenizer([sentence], return_tensors="pt", padding=True, truncation=True).to(device)
    with torch.no_grad():
        outputs = model.generate(**inputs, max_length=128, num_beams=4)
    result = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Show result
    output_text.delete("1.0", tk.END)
    output_text.insert(tk.END, result)


root = tk.Tk()
root.title("English to Arabic Translator")


tk.Label(root, text="Enter your English sentence:").pack()
input_text = tk.Text(root, height=4, width=50)
input_text.pack(pady=5)


translate_button = tk.Button(root, text="Translate to Arabic", command=translate)
translate_button.pack(pady=10)


tk.Label(root, text="Arabic translation:").pack()
output_text = tk.Text(root, height=4, width=50)
output_text.pack(pady=5)


root.mainloop()
```
